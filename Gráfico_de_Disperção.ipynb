## **Plotar o Scatter Plot**

```python
#Importando as bibliotecas pandas e matplotlib do Python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# importar dados para DataFrame
df = pd.read_csv("https://raw.githubusercontent.com/carlosfab/data_science/master/datasets/housing.csv")

# ver as 5 primeiras entradas
df.head()
```

```python
#Plotar o Scatter Plot
fig, ax = plt.subplots(figsize=(10,6))

#Informando quais dados serão plotados
df.plot.scatter(x='median_income',y='median_house_value', ax=ax, s=1, alpha=0.4)

#Mostrando o gráfico
plt.tight_layout();
```

## **Regressão Linear em um Scatter Plot**

```python
#importando o seaborn
import seaborn as sns
```

```python
#Plotar regressão linear
fig, ax = plt.subplots(figsize=(10,6))

#Informando quais dados serão plotados
ax.set_xlim(right=8)
sns.regplot('median_income','median_house_value', scatter_kws={"s": 0.2}, line_kws={"color": "red"}, data=df, ax=ax)

#Mostrando o gráfico
plt.tight_layout()
```
