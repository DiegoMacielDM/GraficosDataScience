# **Trabalhando com gráficos**

```python
#Imporntando as bibliotecas pandas e matplotlib do Python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#importando o DataFrame
df = pd.read_csv("https://raw.githubusercontent.com/carlosfab/curso_data_science_na_pratica/master/modulo_03/dengue-dataset.csv",  index_col="data", parse_dates=True)
#Mostrando o DataFrame
df.head()
```

```python
#Plotar o números de casos confirmados
fig, ax = plt.subplots()
#Informando quais dados serão plotados
df["casos-confirmados"].plot(kind="line", ax=ax)
#Colocando o titulo
ax.set_title("Dengue em Campinas")
#Colocando o titulo valor X
ax.set_xlabel("Data")
#Colocando o titulo valor Y
ax.set_ylabel("Casos confirmados")
#Mostrando o gráfico
plt.tight_layout();
```

## **Selecionando por intervalo**

```python
#Plotar com intevalos
fig, ax = plt.subplots()
#Informando quais dados serão plotados
df.loc['2011-01-01':'2013-12-31', 'temperatura-media'].plot(kind="line", ax=ax)
#Colocando o titulo
ax.set_title("Gráfico da temperatura média")
#Colocando o titulo valor X
ax.set_xlabel("Data")
#Colocando o titulo valor Y
ax.set_ylabel("Temperatura média")
#Mostrando o gráfico
plt.tight_layout();
```

## **PLotar múltiplos gráficos separadamente**

```python
#Plotar vários gráficos
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(10,4))

## Plotar temperatura média
df.loc['2011-01-01':'2013-12-31', 'temperatura-media'].plot(kind="line", ax=ax1)
ax1.set_title("Gráfico da temperatura média")
ax1.set_xlabel("Data")
ax1.set_ylabel("Temperatura média")

## Plotar casos confirmados
df["casos-confirmados"].plot(kind="line", ax=ax2)
ax2.set_title("Casos confirmados")
ax2.set_xlabel("Data")
ax2.set_ylabel("Casos confirmados")

#Mostrando o gráfico
plt.tight_layout();
```

## **Plotar vários gráficos subrepostos**

```python
#Normalizando o plot
df_norm = (df - df.mean())/df.std()
fig, ax = plt.subplots(figsize=(8,4))
## Plotar 1 gráfico - temperatura média
df_norm.loc['2011-01-01':'2013-12-31', 'temperatura-media'].plot(kind="line", ax=ax)

## Plotar 2 gráfico - casos confirmados
df_norm.loc['2011-01-01':'2013-12-31', 'casos-confirmados'].plot(kind="line", ax=ax)

## Plotar casos confirmados
ax.set_title("Temperatura média X Casos confirmados")
ax.set_xlabel("Data")

#Mostrando o gráfico
plt.tight_layout();
```
