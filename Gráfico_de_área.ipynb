```python
#Importando as bibliotecas pandas e matplotlib do Python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# importar dados para DataFrame
df = pd.read_csv("https://raw.githubusercontent.com/carlosfab/curso_data_science_na_pratica/master/modulo_03/dengue-dataset.csv", index_col="data", parse_dates=True)

# ver as 5 primeiras entradas
df.head()
```

```python
#Plotar o histograma da temperatura média
fig, ax = plt.subplots(figsize=(10,6))

#Informando quais dados serão plotados
df.loc['1999-01-01':'2001-01-01', ['temperatura-mininima','temperatura-media','temperatura-maxima']].plot.area(ax=ax)

#Mostrando o gráfico
plt.tight_layout();
```

```python
#PLotar temperatura média X casos confirmados
df_norm = (df-df.mean())/df.std()
```

```python
#Plotar o histograma da temperatura média
fig, ax = plt.subplots(figsize=(10,6))

#Informando quais dados serão plotados
df.loc['1999-01-01':'2001-01-01', ['casos-confirmados','temperatura-media']].plot.area(ax=ax)

#Mostrando o gráfico
plt.tight_layout();
```
