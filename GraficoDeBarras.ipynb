```python
#Importando as bibliotecas pandas e matplotlib do Python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# importar dados para DataFrame
df = pd.read_csv("https://raw.githubusercontent.com/carlosfab/curso_data_science_na_pratica/master/modulo_03/entrada_passageiro_linha1.csv")
```

```python
# ver as 5 primeiras entradas
df.head()
```

```python
#Plotar o gráfico de barras com a entrada média de pessoas
fig, ax = plt.subplots(figsize=(16,8))

#Informando quais dados serão plotados
df.plot(x="Estação", y="Entradas", ax=ax, kind="bar");

#Colocando o titulo 
ax.set_title("Entrada média de passageiros - Linha 1 2018")
#Colocando o titulo de X
ax.set_xlabel("Estação")
#Colocando o titulo de Y
ax.set_ylabel("Entradas")

#Mostrando o gráfico
plt.tight_layout();
```
