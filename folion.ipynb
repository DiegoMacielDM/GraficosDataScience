## **Mapeando dados com Folion**

```python
#Importando as bibliotecas folium e matplotlib do Python
import pandas as pd
import folium
```

```python
mapa = folium.Map(location=[-8.051894,-34.9465504], zoom_start=11)
mapa
```

```python
# importar dados para DataFrame
df = pd.read_csv("/content/drive/MyDrive/estacoesbike.csv",delimiter=';', sep='\t')
```

```python

# ver as 5 primeiras entradas
df.head()
```

```python
#Extraindo os valores do dataframe para variaveis
latitude = df.latitude.values
logintude = df.longitude.values
nomes = df.nome.values

#Criando o mapa
mapa = folium.Map(location=[-8.051894,-34.9465504], zoom_start=11)

#Adicionando pontos ao (makers) ao mapa
for lat, lon, nomes in zip(latitude, logintude, nomes):
	folium.Marker(location=[lat, lon], popup=nomes).add_to(mapa)

#exibindo o mapa final com os pontos
mapa
```
